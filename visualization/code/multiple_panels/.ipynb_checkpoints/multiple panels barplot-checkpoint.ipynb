```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.DataFrame(np.random.rand(3,2))
data['feature'] = ['f%s'%i for i in range(3)]
data.iloc[:, :] = data.values[:, ::-1]
data.columns = ['feature', 'pcc', 'src']
```

```python

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
plt.subplots_adjust(wspace=0)
# add a title for all the panels
plt.suptitle("2019")

n = data.shape[0]

colors = ["#FEAF2D", "#3FA7DA"]
xlabels = ["raw feature", 'contribution']
titles = ["PCC", "SRC"]

for i, ax in enumerate(axes):
    ax.barh(range(n), data.iloc[:, i+1], color=colors[i], alpha=0.7)
    _ = ax.set_yticks(range(n))
    if i == 0:
        _ = ax.set_yticklabels(data.feature)
    else:
        # no ytick and the y axis is gray
        ax.set_yticklabels([])
        ax.tick_params(axis='y', colors='white')
        ax.spines['left'].set_color('gray')
    ax.set_xlabel(xlabels[i])
    ax.set_title(titles[i])
    ax.grid(linestyle='--', linewidth=0.8)

```
