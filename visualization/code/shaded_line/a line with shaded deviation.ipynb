```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array(range(5))
y = np.array(range(5))
up_bound = y + np.random.rand(5)
lower_bound = y-np.random.rand(5)*5

fig, ax = plt.subplots()

l1 = ax.plot(x,y)
ax.fill_between(x, up_bound, lower_bound, alpha=0.2)#, edgecolor=r, facecolor=b)
plt.show()
```
