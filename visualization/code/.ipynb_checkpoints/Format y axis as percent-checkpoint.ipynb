```python
import pandas as pd
import numpy as np
import matplotlib
%matplotlib inline

df = pd.DataFrame(np.random.randn(100,2))

# you get ax from here
ax = df.plot()
type(ax)  # matplotlib.axes._subplots.AxesSubplot

# manipulate
vals = ax.get_yticks()
ax.set_yticklabels(['{:3.2f}%'.format(x*100) for x in vals])
```
