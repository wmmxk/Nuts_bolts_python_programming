```python
# this template shows how to stack three panesl vertically.
# all the panels share  the x axis. 
```

```python
import matplotlib.pyplot as plt
```

```python
def plot_one_ax(ax, label="MI",n=5):
    
    ax.plot(range(n), range(n), label = label)
    ax.set_xlabel("Number of Iteration")
    ax.set_ylabel("Mean absolute error")
    
    ax.legend(loc=2)
```

```python
# create a figure
fig = plt.figure(figsize=(6, 10))
# set the space between panels

# if the hspace is 0, then the only bottom axis shows up
fig.subplots_adjust(wspace=0, hspace=0.3)

# create the first plot
ax = plt.subplot(311)
plot_one_ax(ax, "MI")

# create the 2nd plot
ax = plt.subplot(312, sharex = ax)
plot_one_ax(ax, "EN",n=10)

# create the 3rd plot
ax = plt.subplot(313, sharex = ax)
plot_one_ax(ax, "Tan",n=3)
plt.show()
#plt.savefig("../out_fig/stack_3_panel.pdf", bbox_inches='tight')
```
