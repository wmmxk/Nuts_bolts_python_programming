```python
import matplotlib.pyplot as plt
dct = {'list_1' : [1,2,4,3,1],'list_2' : [2,4,5,1,2],'list_3' : [1,1,3,4,6],'list_4' : [1,1,2,2,1]}

xs = [0,1,2,3,4]


for i in [1,2,3,4]:
    plt.plot(xs,dct['list_%s' %i], label='%s data' % i)


plt.legend(loc = (0.7,0.7))

plt.show()
```

```python
# you can alsor reverse the order of the legends:
import matplotlib.pyplot as plt
%matplotlib inline
dct = {'list_1' : [1,2,4,3,1],'list_2' : [2,4,5,1,2],'list_3' : [1,1,3,4,6],'list_4' : [1,1,2,2,1]}

fig = plt.figure(figsize=(5,4))


ax = fig.add_subplot(111)

for i in [1,2,3,4]:
    plt.plot(xs,dct['list_%s' %i], label='%s data' % i)


handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[::-1], labels[::-1],loc = (0.7,0.7))

#plt.show()

```
