```python
import numpy as np
num = 30 
arr = np.random.rand(num,2)
```

```python
colors = ["#9FCBA6","#CB78FC","#FDA805","#5D99DA"]

SMALL_SIZE = 11
BIGGER_SIZE = 13
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.rc('axes', labelsize= BIGGER_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize


fig = plt.figure(figsize=(5,4))

ax1 = plt.gca()
ax2 = ax1.twinx()
ax1.plot(range(num), arr[:,0] ,color=colors[1],label = "y1")
ax1.set_xlabel('Episodes')
ax1.set_ylabel('Precentage')

L = ax1.legend(bbox_to_anchor=(-0.05, 1.08), loc=3, borderaxespad=1.2)
L.draw_frame(False)

ax2.plot(range(num),arr[:,1],color=colors[2],label="y2")
ax2.set_ylim(-1,1)
L = ax2.legend(bbox_to_anchor=(-0.05, 1.0), loc=3, borderaxespad=1.2)
L.draw_frame(False)


fig.savefig("../out_fig/multiple_axises.pdf", bbox_inches='tight')
```
