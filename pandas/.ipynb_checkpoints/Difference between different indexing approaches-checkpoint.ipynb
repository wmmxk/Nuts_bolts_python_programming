```python
import pandas as pd
import numpy as np
np.random.seed(5)
df = pd.DataFrame(np.random.randint(100, size=(7, 6)), 
                  columns=list('ABCDEF'), 
                  index=['R{}'.format(i) for i in range(7)])
df.head()
```

```python
# swap two columns
```

```python
df[['B', 'A']] = df[['A', 'B']]
df
```

```python
# source: http://pandas.pydata.org/pandas-docs/stable/indexing.html#different-choices-for-indexing
```

```python
#Warning pandas aligns all AXES when setting Series and DataFrame from .loc, .iloc and .ix.
This will not modify df because the column alignment is before value assignment.
```

```python
df.loc[:,['B', 'A']] = df[['A', 'B']]
df
```

```python
df.loc[:,['B','A']] = df[['B','A']]
df
```

```python
# the correct way is to use raw values
df.loc[:,['B', 'A']] = df[['A', 'B']].values
df
```
