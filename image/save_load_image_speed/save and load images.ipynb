```python
import cv2
import time
import numpy as np
start = time.clock()
for i in range(20):
    img = cv2.imread("./Screenshot_11.jpg")
print(time.clock() - start )
np.save("Screenshot_11.npy",img)
print(img.shape)
start = time.clock()
for i in range(20):
    img = np.load("./Screenshot_11.npy")
print(time.clock() - start )

img.tofile("Screenshot_11.dat")

start = time.clock()
for i in range(20):
    fd = open('Screenshot_11.dat','rb')
    img = np.fromfile(file = fd, dtype = np.uint8).reshape(468,494,3)
print(time.clock() - start )

print(img.shape)
```
