```python
x = range(10)
y = [i*2 for i in x]
z = [i*3 for i in x]
q = [i*1.5 for i in x]
```

```python
colors = ["#F47105","#1BACFF","#FDA805"]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.figure(figsize=(5,4))


ax = fig.add_subplot(111)
fig.subplots_adjust(top=0.9)
fig.subplots_adjust(bottom=0.16)
fig.subplots_adjust(left=0.16)

ax.set_xlabel('Episodes')
ax.set_ylabel('Accuracy')


plt.plot(x,y,color=colors[0],label=x)

plt.plot(x,z,color=colors[1],label=x)

plt.plot(x,q,color=colors[2],label=x)

# plt.xticks(np.arange(0, max(episodes[0])+1, 50000))

L = plt.legend()


L.get_frame().set_facecolor('none')

L.get_texts()[0].set_text('Request ratio')
L.get_texts()[1].set_text('Gap between two accuracy')


axes = plt.gca()

axes.set_ylim([0,23])


plt.show()
```
